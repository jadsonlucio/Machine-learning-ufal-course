### Context
The datasets provided include the players data for the Career Mode from FIFA 15 to FIFA 20 ("players_20.csv"). The data allows multiple comparison of the same players across the last 6 version of the videogame.

Some ideas of possible analysis:

Historical comparison between Messi and Ronaldo (what skill attributes changed the most during time - compared to real-life stats);

Ideal budget to create a competitive team (at the level of top n teams in Europe) and at which point the budget does not allow to buy significantly better players for the 11-men lineup. An extra is the same comparison with the Potential attribute for the lineup instead of the Overall attribute;

Sample analysis of top n% players (e.g. top 5% of the player) to see if some important attributes as Agility or BallControl or Strength have been popular or not acroos the FIFA versions. An example would be seeing that the top 5% players of FIFA 20 are more fast (higher Acceleration and Agility) compared to FIFA 15. The trend of attributes is also an important indication of how some attributes are necessary for players to win games (a version with more top 5% players with high BallControl stats would indicate that the game is more focused on the technique rather than the physicial aspect).

# Funções

```python
import sys
from itertools import product
from sklearn.neural_network import MLPRegressor

def progressbar(it, prefix="", size=60, file=sys.stdout):
    count = len(it)
    def show(j):
        x = int(size*j/count)
        file.write("%s[%s%s] %i/%i\r" % (prefix, "#"*x, "."*(size-x), j, count))
        file.flush()        
    show(0)
    for i, item in enumerate(it):
        yield item
        show(i+1)
    file.write("\n")
    file.flush()

def set_tick_attr(weight, fontsize):
    def wrapper(tick_label):
        tick_label.set_fontweight(weight)
        tick_label.set_fontsize(fontsize)
    return wrapper

    
def find_best_configuration(trainX, testX, trainY, testY):
    param_distributions = {
        "hidden_layer_sizes": [(100, 100), (100,), (100, 20)],
        "activation": ["identity", "tanh", "relu"],
        "solver": ["adam", "sgd"],
        "max_iter": [1000, 2000, 5000],
        "learning_rate": ["adaptive"]
    }
    
    results = []

    for config in progressbar(list(product(*param_distributions.values()))):
        params = dict(zip(param_distributions.keys(), config))
        model = MLPRegressor(**params)
        model.fit(trainX, trainY)
        score = model.score(testX, testY)

        results.append((params, score))
    
    results.sort(key = lambda ele: ele[1])
    
    return results[-1], results
```

# Pre-processing

```python
from pandas_profiling import ProfileReport

power_levels_attrs = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf','rw', 'lam', 'cam', 'ram', 'lm',
                      'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb',
                      'cb', 'rcb', 'rb']

df_fifa_2020 = pd.read_csv("datasets/players_20.csv")
df_fifa_2020.head(3)
```

## convert player traits and tags into columns (OneHotEncoder)

```python
import operator
from functools import reduce

def convert_player_attr_into_cols(df, attr, prefix):
    serie = df[attr].str.replace(", ", ",").str.split(",")
    l = serie[~serie.isna()].tolist()
    columns = [name for name in set(list(reduce(operator.concat, l)))]
    data = []
    for player_attrs in serie:
        row = {col:0 for col in columns}
        if isinstance(player_attrs, list):
            for attr in player_attrs:
                row[attr] = 1
        data.append(row)
    
    df = pd.DataFrame.from_dict(data)
    df.columns = [prefix+col_name for col_name in df.columns]
    return df

# OneHotEncoder player_traits attribute
df_player_traits = convert_player_attr_into_cols(df_fifa_2020, "player_traits", "Trait ")
df_fifa_2020 = pd.concat([df_fifa_2020, df_player_traits], axis=1)

# OneHotEncoder player_tags attribute
df_player_tags = convert_player_attr_into_cols(df_fifa_2020, "player_tags", "Tag ")
df_fifa_2020 = pd.concat([df_fifa_2020, df_player_tags], axis=1)
```

## Remove invalid rows and columns

```python
# remove Reserve and Substitute players
df_fifa_2020 = df_fifa_2020[~df_fifa_2020["team_position"].isin(["RES", "SUB"])]

# remove rows where player team position is not available 
df_fifa_2020 = df_fifa_2020[~df_fifa_2020["team_position"].isna()]

# remove goalkeeping duplicate attributes
duplicate_attrs = ['goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking',
                   'goalkeeping_positioning', 'goalkeeping_reflexes']
df_fifa_2020.drop(columns=duplicate_attrs, inplace=True)

# remove attributes that not will be used in player ability inference
unused_attributes = ["sofifa_id", "player_url", "long_name", "dob", "club", "real_face", "release_clause_eur",
                     "player_tags", "player_traits", "team_jersey_number", "loaned_from", "joined", "contract_valid_until",
                     "nation_position", "nation_jersey_number"]

df_fifa_2020.drop(columns=unused_attributes, inplace=True)
```

attributes are in the format power+evolution and i will only get de power part

```python
for attr in power_levels_attrs:
    mask = ~df_fifa_2020[attr].isna()
    df_fifa_2020.loc[mask, attr] = df_fifa_2020[mask][attr].map(lambda string: string.split("+")[0])
    df_fifa_2020[attr]= df_fifa_2020[attr].astype(float)
```

## Convert work rate attribute to integer

The work_rate attribute is represented in the form of strings in the following ways: High / High, High / Low, High / Medium, Low / High, Low / Low, Low / Medium, Medium / High, Medium / Low, Medium / Medium . Where the two names separated by bar represent the player's effort to make offensive or defensive plays even when he is out of his position. In this case, the pre-processing will be to divide this attribute into two new ones, atk_work_rate and def_work_rate.

```python
dict_int_work_rate = {
    "Low": 1,
    "Medium": 2,
    "High": 3
}

def get_int_work_rate(pos=0):
    def wrapper(string):
        return dict_int_work_rate[string.split("/")[pos]]
    return wrapper

df_fifa_2020["atk_work_rate"] = df_fifa_2020["work_rate"].map(get_int_work_rate(0))
df_fifa_2020["def_work_rate"] = df_fifa_2020["work_rate"].map(get_int_work_rate(1))
```

## Data normalization
There are some data that work from a scale of 0 to 100 so for this type of attribute the normalization process will be to divide these attributes by 100. For the other attributes, the process will be the standard of normalization.

```python
# output normalization
df_train_y = df_train_y/100
trainY = df_train_y.values
```

```python
# normalization of input data with and without summarization

def norm_df(df):
    norm_attrs = ['age', 'weak_foot', 'atk_work_rate', 'def_work_rate', 'height_cm', 'weight_kg']
    df1 = df.loc[:, set(df.columns) - set(norm_attrs)]/100
    df2 = df.get(norm_attrs)
    df2 = (df2 - df2.min())/(df2.max() - df2.min())
    
    return pd.concat([df1, df2], axis=1, sort=False)

train_x1 = norm_df(df_train_x1).values
train_x2 = norm_df(df_train_x2).values
```

## Training of linear regression, random forest and mlp models

```python
import random
def cross_validation(model_class, X, y, xlabel, ylabel, evaluation_metrics, cv=5, *args, **kwargs):
    models = []
    models_evaluations = []
    model_name = model_class.__name__
    idx = list(range((len(X)//cv) * cv))
    random.shuffle(idx)
    idxs = np.split(np.array(idx), cv)
    evaluations = {}
    i = 0
    for batch_idx in idxs:
        testX = X[batch_idx]
        testY = y[batch_idx]
        trainX = np.concatenate([X[idxs[j]] for j in range(cv) if j!=i])
        trainY = np.concatenate([y[idxs[j]] for j in range(cv) if j!=i])
        
        model = model_class(*args, **kwargs)
        model.fit(trainX, trainY)
        models.append(model)
        
        for evaluation_metric in evaluation_metrics:
            evaluation = evaluation_metric(model.predict(testX), testY)
            models_evaluations.append([xlabel, ylabel, model_name, evaluation_metric.__name__, evaluation])
           
        
        i+=1
    
    return models, models_evaluations
        
```

```python
from itertools import product
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

def custom_mlp_regressor():
    return MLPRegressor(**best_result_basic[0])

models = [LinearRegression, custom_mlp_regressor, RandomForestRegressor]
inputs = [(train_x1, "trainX"), (train_x2, "trainX_summarized")]
data = []
for model, trainX in product(models, inputs):
    trainX, xlabel = trainX
    models, models_results = cross_validation(model, trainX, trainY, xlabel, "trainY", 
                                             [metrics.r2_score, metrics.mean_squared_error])
    data += models_results

df_acurracy = pd.DataFrame(columns=["xlabel", "ylabel", "model", "metric", "value"], data=data)
```

# Evaluation of models

The results with the normal (trainX) and summarized (trainX_summarized) inputs were very similar, to confirm that the two types of input have the same efficiency, a wilcoxon hypothesis test was run using the metrics of r2 and mse (mean_squared_error) comparing the results using both inputs. The result was a p-value of 0.17 for the r2 score and 0.069 using the mse, thus, assuming an alpha of 0.05 (5% error) we have that both p-values are greater than this value, so we can reject the null hypothesis that the acurracias are equal and assume that there is a difference between normal and summarized entries. The model that had the best performance in general (with both inputs), the random forest obtained an accuracy of 96% on average, against 94% of the linear regression and 94% of the mlp.

## Wage attr pre-processing

From the graph below we can see that there are many outliers in the player's salary, which is expected since many players receive astronomical salaries far above the average. Due to the number of outliers common standardization techniques will not work, since they are based on maximum or average values to work. One way to get around this is to use a technique that takes into account the positioning of the data instead of the value itself.

```python
# boxplot player wage
wage = df_players.get(["wage_eur"]).values.ravel()
ax = sns.boxplot(wage, orient='v')
ax.set_ylabel("Player Wage")
```

```python
# remove outliers

scaler = preprocessing.QuantileTransformer(random_state=0)
wage = df_players.get(["wage_eur"])
q1 = wage.quantile(0.25)
q3 = wage.quantile(0.75)
iqr = q3 - q1
mask = ~(((q3 + 1.5 * iqr) < wage) | ((q1 - 1.5 * iqr) > wage))
mask = mask.values.ravel()
wage = wage.values[mask]
ax = sns.boxplot(wage, orient='v')
ax.set_ylabel("Player Wage")
```

## Train with pre-processed wage

The results show a much lower value of r2 in relation to the use without removing the outliers, a possible explanation for this may be the fact that the data variance has been removed together with the outliers, so the model does not have enough divergence in the going out to make a good inference.

```python
# train with train_x1 attrs
trainX1 = np.compress(mask, train_x1, axis=0)
x1_model, x1_r2, x1_mse = train_random_forest(trainX1, wage.ravel(), random_state=10)
# train with train_x2 attrs
trainX2 = np.compress(mask, train_x2, axis=0)
x2_model, x2_r2, x2_mse = train_random_forest(trainX2, wage.ravel(), random_state=10)

fig = px.bar(x=["trainX", "trainX_summarized"], y=[x1_r2, x2_r2])
fig.update_layout(yaxis_title="r2 score", xaxis_title="train data")
Image(fig.to_image())
```
